### 📌  Analysis of Duplicate Response Patterns
1. Creating pattern groups   

```python
group_cols = ["MainBranch", "Employment", "RemoteWork"]

pattern_groups = df.groupby(group_cols).size().reset_index(name="count")
pattern_groups.sort_values("count", ascending=False).head(10)
```

<p style="color:#666; font-size:13px;">
<u><b>Insight 📌 Response Patterns</b></u><br>
The majority of respondents are full-time developers, with hybrid work being the most common working arrangement. This shows a strong concentration of traditional software engineering roles combined with flexible work models.
</p>

2. Identifying strongest pattern group


```python
top_group = df.groupby(group_cols).size().idxmax()
print(top_group)
```

<p style="color:#666; font-size:13px;">
<u><b>Insight 📌 Most Common Response Pattern</b></u><br>
The most frequent combination of responses indicates that the largest group of participants are professional developers working full-time in hybrid environments. This highlights the dominance of flexible work models within the global developer community.
</p>

3. Analyzing similarity inside group


```python
group_country_distribution = (
    df.groupby(group_cols)["Country"]
    .value_counts()
    .reset_index(name="count")
)
group_country_distribution 
group_country_distribution .sort_values("count", ascending=False).head(10)
```

<p style="color:#666; font-size:13px;">
<u><b>Insight 📌 Country Distribution Within Groups</b></u><br>
The analysis shows that within the most common developer and employment patterns, certain countries such as the United States and India appear most frequently. This indicates that these regions have a strong representation in the global developer workforce across different work arrangements.
</p>

4. Visualize Shared Response Patterns

```python
pattern_distribution = (
    df.groupby(group_cols)
        .size()
        .reset_index(name="count")
    .sort_values("count", ascending=False)
)
pattern_distribution.head(10)
```

<p style="color:#666; font-size:13px;">
<u><b>Insight 📌 Shared response patterns</b></u><br>
The dataset reveals that a small number of combinations of MainBranch, Employment, and RemoteWork account for a large share of responses. This indicates that most respondents fall into a few dominant developer profiles, such as full-time developers working in hybrid or remote environments. This concentration highlights common working structures within the global developer community.
</p>

```python
top_patterns = pattern_distribution.head(10)
top_patterns 
```

<p style="color:#666; font-size:13px;">
<b>Insight 📌 Top response patterns</b><br>
The top 10 patterns represent the most frequent combinations of developer role, employment type, and remote work style. This shows that a small number of standardized work arrangements (such as full-time developers in hybrid or remote roles) dominate the dataset, highlighting common global employment structures in the software development industry.
</p>

```python
labels = top_patterns["MainBranch"] + " | " + top_patterns ["Employment"] + " | "  + top_patterns ["RemoteWork"]   
labels
```

<p style="color:#666; font-size:13px;">
<b>Insight 📌 Combined labels for visualization</b><br>
This step creates readable labels by merging MainBranch, Employment, and RemoteWork into a single string for each record in the top patterns dataset. These combined labels are used to improve chart readability, allowing each bar in the visualization to clearly represent a complete response profile instead of separate categorical fields.
</p>

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.barh(labels, top_patterns["count"])

plt.xlabel("Number of Respondents")
plt.title("Top 10 Shared Response Patterns")
plt.gca().invert_yaxis()

plt.show()
```

<p style="color:#666; font-size:13px;">
<b>Insight 📌 Shared response patterns visualization</b><br>
The bar chart highlights the top 10 most common combinations of MainBranch, Employment, and RemoteWork among respondents. It shows that a small number of developer profiles dominate the dataset, with full-time developers working in hybrid and remote environments being the most frequent. This visualization makes it easier to compare the prevalence of each response pattern at a glance.
</p>

### 📌 Evaluation of Duplicates and Handling Strategy

```python
#1. ID-based uniqueness
df["ResponseId"].is_unique
```

```python
# 2. Full-row duplicates
df.duplicated().sum()
```

```python
#3. Logical uniqueness
df.duplicated(subset=["MainBranch", "Employment", "RemoteWork"]).sum()
```

<p style="color:#666; font-size:13px;">
<b>Insight 📌 Duplicate evaluation</b><br>
The ResponseId column is not fully unique and should not be treated as a strict primary key. Only a small number of exact duplicate rows (20) were identified, indicating minimal data quality issues. However, a large number of repeated patterns appear across MainBranch, Employment, and RemoteWork, which is expected in survey data and reflects common developer profiles rather than duplication errors.

Removing rows based on partial-column duplication is not appropriate, as these fields represent valid and frequently occurring behavioral patterns rather than unique identifiers. Such removal would distort the dataset by eliminating meaningful responses and biasing the analysis. Therefore, only exact duplicate rows should be removed, while all other records should be retained to preserve the integrity and analytical value of the dataset.
</p>

### 📌 Verify and Document Duplicate Removal Process


```python
df.duplicated().sum()
```

```python
df[df.duplicated()].head()
```

```python
df_cleaned = df.drop_duplicates()
```

```python
df_cleaned.duplicated().sum()
```

<p style="color:#666; font-size:13px;">
<b>Insight 📌 Duplicate handling process</b><br>
Duplicate records were first identified using the duplicated() function, which revealed a small number of exact duplicate rows in the dataset. These were reviewed to confirm they represented redundant entries rather than meaningful responses. The duplicates were then removed using the drop_duplicates() method to ensure data integrity. After cleaning, the dataset contained only unique records, while maintaining all valid survey response patterns for analysis.
</p>
