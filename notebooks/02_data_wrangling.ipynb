### 📌 Data Wrangling
#####  Identifying and Removing Inconsistencies

```python
def check_column_quality(df, col):
    print(f"\n🔍 Analyzing column: {col}\n")

    # 1. Missing values
    print("❌ Missing values:")
    print(df[col].isnull().sum())

    # 2. Unique values (sample preview)
    print("\n📌 Sample unique values:")
    print(df[col].dropna().unique()[:20])

    # 3. Value counts (top categories)
    print("\n📊 Top value counts:")
    print(df[col].value_counts().head(10))

    # 4. Check whitespace issues
    print("\n⚠️ Values with leading/trailing spaces:")
    print(df[col].dropna().astype(str).str.contains(r"^\s|\s$", regex=True).sum())

    # 5. Check lowercase vs uppercase variation
    print("\n🔠 Case variation (normalized duplicates check):")
    print(df[col].dropna().astype(str).str.lower().nunique())

    # 6. Rare categories (potential noise)
    print("\n📉 Rare values (count <= 5):")
    vc = df[col].value_counts()
    print(vc[vc <= 5])
```

```python
check_column_quality(df, "Country")
```

<p style="color:#666; font-size:13px;">
<b>Insight 📌 Country column data quality analysis</b><br>
The Country column contains a small number of missing values and shows a highly imbalanced distribution, where a few countries dominate the dataset. Additionally, there is a long tail of rare countries with very low frequency and some inconsistencies in country naming conventions. This indicates the need for standardization of country names and grouping of rare categories to ensure more reliable and meaningful analysis.
</p>

```python
# 1. Handle missing first
df["Country"] = df["Country"].fillna("Unknown")

# 2. Clean text
df["Country"] = df["Country"].str.strip()

# 3. Standardize names
df["Country"] = df["Country"].replace({
    "Swaziland": "Eswatini",
    "Iran, Islamic Republic of...": "Iran",
    "Democratic People's Republic of Korea": "North Korea"
})

# 4. Group rare categories
vc = df["Country"].value_counts()
rare = vc[vc <= 5].index
df["Country"] = df["Country"].replace(rare, "Other")

df["Country"].isnull().sum()
```

<p style="color:#666; font-size:13px;">
<b>Insight 📌 Country column data cleaning and standardization</b><br>
The Country column was cleaned by first handling missing values, which were replaced with "Unknown" to preserve dataset completeness. Text formatting issues such as leading or trailing spaces were removed to ensure consistency. Inconsistent country names were standardized (e.g., "Swaziland" to "Eswatini" and variations of Iran and North Korea were unified). Finally, rare countries with very low frequency (≤ 5 occurrences) were grouped into a single category labeled "Other" to reduce noise and improve interpretability. This cleaning process improves data quality and ensures more reliable categorical analysis.
</p>
